```python
#Import the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary_Data.csv')
```

```python
#Create Matrix of features X
X = dataset.iloc[:, :-1].values
```

```python
#Vector of Y dependent variable 
y = dataset.iloc[:,1].values
```

```python
#Spilitting data sets into training and test
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size = 1/3, random_state = 0)
```

```python
# Import Linear Regression fron sklearn and create regressor object
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predicting the Test set results
#Create vector of predicted variable
y_pred = regressor.predict(X_test)
```

```python
#Visualizing  the training results

plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train,regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience(Training set)')
plt.xlabel('Years of Exp')
plt.ylabel('Salary')
plt.show()
```
